```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data = pd.read_csv("Advertising.csv")
```

```python
data.drop(data.columns[0], axis=1, inplace=True)
```

```python
params = ["TV", "radio", "newspaper"]

sales = data['sales']

for i, param in enumerate(params, 1):
    pred = np.array(data[param])
    pred = pred[:, np.newaxis]
    reg = LinearRegression().fit(pred, sales)
    x = np.linspace(0, pred.max(), 10)
    y = reg.intercept_ + reg.coef_ * x
    plt.subplot(1, 3, i)
    plt.scatter(x=pred, y=sales, marker='o', facecolor='none', edgecolors='red')
    plt.plot(x, y)
```
